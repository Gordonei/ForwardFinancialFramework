## Method

Model Verification Procedure:
1. Run each task for up to (5 minutes)/128 on each platform
2. Generate Latency, Accuracy and Latency-Accuracy models for each task.
3. Combine the task models for each platform and make a prediction curve for the region of interest [5x2,5x4,5x8,5x16]
4. Run verification experiments so as to find that region of interest

#### (2) Generating Models

##### Latency Models

Two sets of data need to be selected:
* All of the paths and latency data for a particular task
* All of the paths and latency data under the benchmarking time limit (300S/128) which will be used to create the model

Latency as a function of the number of simulations. This is a simple linear model - a single degree per simulation component and a constant setup component.

```python
def generate_latency_model(paths,latency):
    latency_model_data = numpy.column_stack((paths,numpy.ones(len(paths))))
    latency_model_coefficients = numpy.linalg.lstsq(latency_model_data,latency)[0]
    
    return lambda N: latency_model_coefficients[0]*N+latency_model_coefficients[1]
```

Calculating the relative and absolute errors for the latency model. The error is the verification latency recorded minus the value predicted by the model. An error > 0 means that the model is *underestimating* the simulations required to achieve that latency. An error < 0 means that the model is *overestimating* the latency required. 

```python
relative_latency_errors = []
absolute_latency_errors = []
verification_paths = []

platforms_interest = [(platform,platform_type) for platform,platform_type in platforms[:]] # if ("phi" not in platform and "opencl_gpu" not in platform_type)
for platform,platform_type in platforms_interest:
    print(platform,platform_type)
    relative_latency_errors += [[]]
    absolute_latency_errors += [[]]
    verification_paths += [[]]
    
    benchmark_paths = []
    for t in range(len(option_tasks[:])):
        selection = data["Platform"]==platform
        selection &= data["Type"]==platform_type
        selection &= data["Task"]==str(t)
        verification_selection = numpy.copy(selection)
        selection &= data["Latency"].astype(numpy.float) < 300.0/128 #limiting to the benchmarking time
        
        paths = data[selection]["Paths"].astype(numpy.float)
        latency = data[selection]["Latency"].astype(numpy.float)
        latency_model = generate_latency_model(paths,latency)

        verification_path = data[verification_selection]["Paths"].astype(numpy.float)
        verification_latency = data[verification_selection]["Latency"].astype(numpy.float)
        
        error = (numpy.array(verification_latency-numpy.array([latency_model(p) for p in verification_path])))
        if not(numpy.any(numpy.isnan(error))or numpy.any(numpy.isinf(error))):
            benchmark_paths += [numpy.max(paths)]
            verification_paths[-1] += [verification_path]
            absolute_latency_errors[-1] += [error]
            relative_latency_errors[-1] += [error/verification_latency]
    
    #Find the number of benchmark paths used across the tasks
    benchmark_paths = numpy.min(benchmark_paths)
    
    #massage error arrays down to same size
    max_length = min([len(errors) for errors in relative_latency_errors[-1]])
    
    temp_errors = relative_latency_errors[-1][:]
    relative_latency_errors[-1] = numpy.array([te[:max_length] for te in temp_errors])
    
    temp_paths = verification_paths[-1][:]
    verification_paths[-1] = numpy.array([tp[:max_length]/benchmark_paths for tp in temp_paths])
    
    temp_errors = absolute_latency_errors[-1][:]
    absolute_latency_errors[-1] = numpy.array([te[:max_length] for te in temp_errors])
    
    print(verification_paths[-1].shape,relative_latency_errors[-1].shape,absolute_latency_errors[-1].shape)
```

The relative latency error across the platforms shows a fair amount of variation in the benchmark region, but relatively speaking is surprisingly low outside of it. An explanation for this is that the primary source of error in latency model is in the constant setup component of the model, which is suspetable to multiple sources of noise (OS, network probably being the main culprits).

```python
f, (ax1, ax2) = plt.subplots(2,sharex=True)

#Top plot
numpy.random.seed(1234)
for i,(platform,platform_type) in enumerate(platforms_interest):
    ax1.plot(verification_paths[i][0],100*numpy.mean(relative_latency_errors[i],axis=0),"-x",label=platform+"("+ platform_type+")") #paths should all be the same
    
ax1.axvspan(1, 1,facecolor='g', alpha=1)

#ax1.set_xlabel("Runtime to Benchmark time Ratio")
ax1.set_xscale('log')
ax1.set_ylabel("Relative Latency Error (%)")
#ax1.set_yscale('log')
ax1.set_ylim((-20,20))

#Bottom plot
numpy.random.seed(1234)
for i,(platform,platform_type) in enumerate(platforms_interest):
    ax2.plot(verification_paths[i][0],numpy.abs(numpy.mean(absolute_latency_errors[i],axis=0)),"-x",label=platform+"("+ platform_type+")") #paths should all be the same
    
ax2.axvspan(1, 1,facecolor='g', alpha=1)

ax2.set_xlabel("Runtime to Benchmark time Ratio")
ax2.set_xscale('log')
ax2.set_ylabel("Absolute Latency Error (S)")
#ax2.set_ylim((-0.1,0.1))
ax2.set_yscale('log')

fig = plt.gcf()
fig.set_size_inches(18.5,21)

plt.legend(loc="best")
```

##### Accuracy Model

Function for generating accuracy model

```python
def generate_accuracy_model(paths,accuracy):
    accuracy_model_data = numpy.column_stack((paths**-0.5,numpy.zeros(len(paths))))
    accuracy_model_coefficients = numpy.linalg.lstsq(accuracy_model_data,accuracy)[0]
    
    return lambda N,accuracy_coefficient=accuracy_model_coefficients[0]: accuracy_coefficient*N**-0.5
```

Calculating the relative and absolute accuracy models. The error is the actual 95% confidence interval measured versus what is the 95% confidence interval predicted by the model for that number of simulations. An error > 0 indicates that the model has *overestimated* the number of simulations required to achieve that level of accuracy, while an error < 0 indicates that the model has *underestimated* the simulations required.

The relative error performance will be limited by any inherent bias in the model.

```python
relative_accuracy_errors = []
absolute_accuracy_errors = []
verification_paths = []

platforms_interest = [(platform,platform_type) for platform,platform_type in platforms[:]]# if ("phi" in platform and "opencl_gpu" in platform_type)]
for platform,platform_type in platforms_interest:
    print(platform,platform_type)
    absolute_accuracy_errors += [[]]
    relative_accuracy_errors += [[]]
    verification_paths += [[]]
    
    benchmark_paths = []
    for t in range(len(option_tasks[:])):
        selection = data["Platform"]==platform
        selection &= data["Type"]==platform_type
        selection &= data["Task"]==str(t)
        verification_selection = numpy.copy(selection)
        selection &= data["Latency"].astype(numpy.float) < 300.0/128 #limiting to the benchmarking time
        
        paths = data[selection]["Paths"].astype(numpy.float)
        accuracy = data[selection]["Accuracy"].astype(numpy.float)
        accuracy_model = generate_accuracy_model(paths,accuracy)

        verification_path = data[verification_selection]["Paths"].astype(numpy.float)
        verification_accuracy = data[verification_selection]["Accuracy"].astype(numpy.float)
        
        error = (numpy.array(verification_accuracy-numpy.array([accuracy_model(p) for p in verification_path])))
        
        #print(error)
        if not(numpy.any(numpy.isnan(error))or numpy.any(numpy.isinf(error))):
            benchmark_paths += [max(paths)]
            verification_paths[-1] += [verification_path]
            absolute_accuracy_errors[-1] += [error]
        else: 
            print("no absolute error:",t)
            print("nan",numpy.any(numpy.isnan(error)))
            print("inf",numpy.any(numpy.isinf(error)))
        
        if not(numpy.all(verification_accuracy<=0.0) or numpy.any(numpy.isnan(error))):
            relative_accuracy_errors[-1] += [[]]
            for e,va in zip(error,verification_accuracy):
                if(va>0): relative_accuracy_errors[-1][-1] += [e/va]
                else: relative_accuracy_errors[-1][-1] += [0.0]
                    
            relative_accuracy_errors[-1][-1] = numpy.array(relative_accuracy_errors[-1][-1])
        else: 
            print("no relative error",t)
            print("zero error",not(numpy.any(error==0.0)))
            print("verification zero",not(numpy.any(verification_accuracy==0.0)))
        
            
    #Find the number of benchmark paths used across the tasks
    benchmark_paths = numpy.min(benchmark_paths)
    
    #massage down to same size
    max_length = min([len(errors) for errors in relative_accuracy_errors[-1]])
    
    temp_errors = relative_accuracy_errors[-1][:]
    relative_accuracy_errors[-1] = numpy.array([te[:max_length] for te in temp_errors])
    
    temp_paths = verification_paths[-1][:]
    verification_paths[-1] = numpy.array([tp[:max_length]/benchmark_paths for tp in temp_paths])
    
    temp_errors = absolute_accuracy_errors[-1][:]
    absolute_accuracy_errors[-1] = numpy.array([te[:max_length] for te in temp_errors])
    
    print(verification_paths[-1].shape,relative_accuracy_errors[-1].shape,absolute_accuracy_errors[-1].shape)
```

Function for generating a Latency as a function of accuracy model. It could actually be done without the paths variable, which is just used to link the latency and accuracy.

What's quite important is that any model that is returning a nonsensical value, either implicity as a 0 coefficient or a nan is discounted. This isn't very Pythonic, but makes sense in the context of best-effort interfaces.

```python
def latency_accuracy_model_generate(paths,latency,accuracy):
    latency_model_data = numpy.column_stack((paths,numpy.ones(len(paths))))
    latency_model_coefficients = numpy.linalg.lstsq(latency_model_data,latency)[0]
    #latency_model = lambda N: latency_model_coefficients[0]*N + latency_model_coefficients[1]

    accuracy_model_data = numpy.column_stack((paths**-0.5,numpy.zeros(len(paths))))
    accuracy_model_coefficients = numpy.linalg.lstsq(accuracy_model_data,accuracy)[0]
    #accuracy_model = lambda N: accuracy_model_coefficients[0]*N**-0.5 + accuracy_model_coefficients[1]
            
    if(accuracy_model_coefficients[0]<=0 or numpy.isnan(accuracy_model_coefficients[0])): 
        return lambda A,latency_coefficient=latency_model_coefficients[0],accuracy_coefficient=accuracy_model_coefficients[0],latency_setup=latency_model_coefficients[1]: 0.0
    else: 
        return lambda A,latency_coefficient=latency_model_coefficients[0],accuracy_coefficient=accuracy_model_coefficients[0],latency_setup=latency_model_coefficients[1]: latency_coefficient*(accuracy_coefficient/A)**2 + latency_setup 
```

The big issue with the plot below is in comparing model curves and the verification points. Each point should be the time taken to achieve that accuracy for all of the tasks. For the predictive models this is easy to do:
1. The latency accuracy model is generated from the benchmark data, i.e. all data for which the time 300.0/number of tasks
2. The curve is plotted according to the domain of interest
These are the dashed curves in the plot.

However, this isn't as straight-forward for the verification data. The verification runs aren't that fine-grained due to the multiplicative stepping. So I've done two things:
1. Simply plot the latency-accuracy model using all of the available data as opposed to just the benchmark data. This is the solid line plot.
2. Find the accuracies achieved that are below each accuracy point of interest (as defined in the accuracy range), and add the minimum of these latencies to the total for that point of interest. Only those accuracy points which have some high fraction of all tasks are plotted (currently 0.9). This is the scatter plot points.

I should probably do more fine-grained verification runs.

```python
fig, ax1 = plt.subplots(1)

numpy.random.seed(1234)
ax1.axvspan(300, 300, alpha=0.5)
#ax1.axvspan(3000, 3000, alpha=0.5)

platforms_interest = [(platform,platform_type) for platform,platform_type in platforms] #if "cpu" in platform_type or "fpga" in platform_type
for platform,platform_type in platforms_interest:
    latency_accuracy_models = []
    verification_latency_accuracy_models = []
    
    verification_latencies = numpy.zeros(accuracy_range.shape)
    verification_sets = [set() for i in verification_latencies]
    for task,dummy in enumerate(option_tasks[:]):
        selection = data["Platform"]==platform
        selection &= data["Type"]==platform_type
        selection &= data["Task"].astype(numpy.int)==task 
        verification_selection = numpy.copy(selection)
        #selection &= data["Latency"].astype(numpy.float)<300.0/128 #constrain the benchmarking
        
        if(numpy.any(selection)):
            paths = data[selection]["Paths"].astype(numpy.float)
            accuracy = data[selection]["Accuracy"].astype(numpy.float)
            latency = data[selection]["Latency"].astype(numpy.float)
            
            latency_accuracy_models += [latency_accuracy_model_generate(paths,latency,accuracy)]
            
            verification_paths = data[verification_selection]["Paths"].astype(numpy.float)
            verification_accuracy = data[verification_selection]["Accuracy"].astype(numpy.float)
            verification_latency = data[verification_selection]["Latency"].astype(numpy.float)
            
            verification_latency_accuracy_models += [latency_accuracy_model_generate(verification_paths,verification_latency,verification_accuracy)]
            
            #plt.scatter(verification_latency,verification_accuracy,marker="x") #plots all of the task points
            #finding the closest accuracy in the range of interest and accumulating the accuracy
            
            for i,acc in enumerate(accuracy_range):
                accuracy_selection = (verification_accuracy-acc)<=0 #selecting those entries at the required accuracy or lower
                if(numpy.any(accuracy_selection)): 
                    verification_latencies[i] += min(verification_latency[accuracy_selection])
                    verification_sets[i] |= {task}
                
            #for acc,lat in zip(verification_accuracy,verification_latency):
                #index = numpy.argmin([(acc-accuracy_range)<0])
                #verification_latencies[index] += lat
                #
            
    #Building the model of models
    latency_accuracy_model = lambda A: numpy.sum([lam(A) for lam in latency_accuracy_models])
    verification_latency_accuracy_model = lambda A: numpy.sum([lam(A) for lam in verification_latency_accuracy_models])
    
    colour = tuple(numpy.random.random(3))
    plot = ax1.plot([latency_accuracy_model(a) for a in accuracy_range],accuracy_range,"--",label=platform+" ("+ platform_type+")",color=colour)
    
    plot_latencies = []
    plot_accuracies = []
    for lat,acc,vset in zip(verification_latencies,accuracy_range,verification_sets):
        #if( 30 < latency_accuracy_model(acc)< 300 ):
        if(len(vset)>=len(option_tasks)):
            plot_latencies += [lat]
            plot_accuracies += [acc]
    
    plot = ax1.plot([verification_latency_accuracy_model(a) for a in accuracy_range],accuracy_range,"-",color=colour)
    plt.scatter(plot_latencies,plot_accuracies,color=colour)
            
ax1.set_xscale('log')
ax1.set_xlabel("Latency (S)")
ax1.set_yscale('log')
ax1.set_ylabel("Accuracy ($)")
ax1.legend(loc='best')

fig.set_size_inches(18.5,10.5)
```

For using the data in the partitioners, the coefficients for the models are required for each task upon each platform

```python
def latency_accuracy_model_coefficients(paths,latency,accuracy):
    setup_ones = numpy.ones(len(paths))
    latency_model_data = numpy.column_stack((paths,setup_ones))
    latency_model_coefficients = numpy.linalg.lstsq(latency_model_data,latency)[0]
    
    selection = latency_model_coefficients[1]<0
    if selection:
        setup_zeros = numpy.zeros(len(paths))
        latency_model_data = numpy.column_stack((paths,setup_zeros))
        latency_model_coefficients = numpy.linalg.lstsq(latency_model_data,latency)[0]

    accuracy_model_data = numpy.column_stack((paths**-0.5,numpy.zeros(len(paths))))
    accuracy_model_coefficients = numpy.linalg.lstsq(accuracy_model_data,accuracy)[0]
    #accuracy_model = lambda N: accuracy_model_coefficients[0]*N**-0.5 + accuracy_model_coefficients[1]
            
    if(numpy.any(latency_model_coefficients<0) or accuracy_model_coefficients[0]<0 or numpy.isnan(accuracy_model_coefficients[0])): 
        raise ValueError("Model Values invalid: %s"%str((accuracy_model_coefficients[0],latency_model_coefficients[0],latency_model_coefficients[1])))
        #return (numpy.inf,numpy.inf,numpy.inf)
    else: return (accuracy_model_coefficients[0],latency_model_coefficients[0],latency_model_coefficients[1])
```
