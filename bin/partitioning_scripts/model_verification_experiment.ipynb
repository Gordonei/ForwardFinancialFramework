## Method

Model Verification Procedure:
1. Run each task for up to (5 minutes)/128 on each platform
2. Generate Latency, Accuracy and Latency-Accuracy models for each task.
3. Combine the task models for each platform and make a prediction curve for the region of interest [5x2,5x4,5x8,5x16]
4. Run verification experiments so as to find that region of interest

```python
with open('benchmark_data_complete_seed4_13000.csv','r') as benchmark_file:
    csvreader = csv.reader(benchmark_file,delimiter=',')
    
    headings = csvreader.next()
    heading_dictionary = {"Latency":("Latency","f"),"Accuracy":("Accuracy","f"),"Paths":("Paths","d")}
    heading_tuples = [heading_dictionary.get(h,(h,"a30")) for h in headings]
    
    data = []
    for row in csvreader: data.append(tuple(row))
            
    data = numpy.array(data,dtype=heading_tuples)
```

Function for selecting particular data entries

```python
def data_select(data,platform_name,platform_type,task,latency_limit=None,accuracy_limit=None):
    selection = data["Platform"]==platform_name
    selection &= data["Type"]==platform_type
    selection &= data["Task"]==task
    
    if(latency_limit): selection &= data["Latency"] <= latency_limit
    if(accuracy_limit): selection &= data["Accuracy"] <= accuracy_limit
    
    return numpy.copy(data[selection])
```

```python
def task_data_select(data,task,latency_limit=None,accuracy_limit=None):
    selection = data["Task"]==task
    
    if(latency_limit): selection &= data["Latency"] <= latency_limit
    if(accuracy_limit): selection &= data["Accuracy"] <= accuracy_limit
    
    return numpy.copy(data[selection])
```

#### (2) Generating Models

##### Latency Models

Two sets of data need to be selected:
* All of the paths and latency data for a particular task
* All of the paths and latency data under the benchmarking time limit (300S/128) which will be used to create the model

Latency as a function of the number of simulations. This is a simple linear model - a single degree per simulation component and a constant setup component.

```python
def generate_latency_model(paths,latency):
    #if(len(latency)<2): raise Exception("Undetermined system")
    
    #latency_model_data = numpy.column_stack((paths,numpy.ones(len(paths))))
    #latency_model_coefficients = numpy.linalg.lstsq(latency_model_data,latency)[0]
    
    #weights = (paths/numpy.min(paths))**0.25
    latency_func = lambda x,latency_coefficient,latency_offset: latency_coefficient*x + latency_offset
    latency_model_coefficients = scipy.optimize.curve_fit(latency_func,paths,latency)[0] #,sigma=weights
    
    #if any of the coefficients are less than zero, first try reduce the data points, then try without a setup latency
    if(numpy.any(latency_model_coefficients[:1]<0)): #raise Exception("Latency model coefficients should not be less than 0!")
        offset = 1
        while(numpy.any(latency_model_coefficients[:1]<0) and (len(paths)-offset)>=2):
            latency_func = lambda x,latency_coefficient,latency_offset: latency_coefficient*x + latency_offset
            latency_model_coefficients = scipy.optimize.curve_fit(latency_func,paths[offset:],latency[offset:])[0] #,sigma=weights[offset:]
            #latency_model_data = numpy.column_stack((paths[offset:],numpy.ones(len(paths[offset:]))))
            #latency_model_coefficients = numpy.linalg.lstsq(latency_model_data,latency[offset:])[0]
            offset += 1
        
        if(numpy.any(latency_model_coefficients[:1]<0)):
            latency_func = lambda x,latency_coefficient,latency_offset: latency_coefficient*x #+ latency_offset
            latency_model_coefficients = scipy.optimize.curve_fit(latency_func,paths,latency)[0] #,sigma=weights
            #latency_model_data = numpy.column_stack((paths,numpy.zeros(len(paths))))
            #latency_model_coefficients = numpy.linalg.lstsq(latency_model_data,latency)[0]
            latency_model_coefficients[1] = 0
        
    elif(latency_model_coefficients[1]<0): #raise Exception("Latency model coefficients should not be less than 0!")
        latency_model_coefficients[1] = 0.0
    
    return lambda N,lmc=latency_model_coefficients[0],lms=latency_model_coefficients[1]: lmc*N+lms
```

Generating data for culmulative error plot. For a constant runtime target, capturing error as a function of the benchmarking time. 

The error is the verification latency recorded minus the value predicted by the model. An error > 0 means that the model is *underestimating* the simulations required to achieve that latency. An error < 0 means that the model is *overestimating* the latency required. 

```python
datapoints = 6
platforms_interest = platforms[:]
tasks_interest = option_tasks[:]

benchmark_runtime_paths_ratio = numpy.empty((len(platforms_interest),datapoints))
benchmark_runtime_latency_ratio = numpy.empty_like(benchmark_runtime_paths_ratio)
relative_latency_error = numpy.empty_like(benchmark_runtime_paths_ratio)
absolute_latency_error = numpy.empty_like(benchmark_runtime_paths_ratio)

for p,(platform,platform_type) in enumerate(platforms_interest):
    #The Xeon Phi needs an exception added
    #if(platform!="phi" and platform_type!="opencl_gpu"): benchmarktime_threshold = 600.0/128
    #else: benchmarktime_threshold = 1280.0/128

    task_benchmark_runtime_paths_ratio = numpy.empty((len(tasks_interest),datapoints))
    task_benchmark_runtime_latency_ratio = numpy.empty_like(task_benchmark_runtime_paths_ratio)
    task_relative_error = numpy.empty_like(task_benchmark_runtime_paths_ratio)
    task_absolute_error = numpy.empty_like(task_benchmark_runtime_paths_ratio)
    for t in range(len(tasks_interest)):
        task = str(t)
        complete_benchmark_data = data_select(data,platform,platform_type,task) #,latency_limit=fixed_runtime
        task_datapoints = complete_benchmark_data.shape[0]
        if(task_datapoints<datapoints): raise Exception("Insufficient datapoints for task %d on %s(%s) - only %d/%d available"%(t,platform,platform_type,task_datapoints,datapoints))
        
        fixed_runtime_paths = complete_benchmark_data["Paths"][datapoints] #[-1]
        fixed_runtime_latency = complete_benchmark_data["Latency"][datapoints] #[-1]
        
        for i,benchmark_datapoint in enumerate(reversed(complete_benchmark_data[1:datapoints+1])):
            task_benchmark_runtime_paths_ratio[t,i] = numpy.copy(1.0*benchmark_datapoint["Paths"]/fixed_runtime_paths) #fixed_runtime_latency
            task_benchmark_runtime_latency_ratio[t,i] = numpy.copy(1.0*benchmark_datapoint["Latency"]/fixed_runtime_latency)
            
            #selecting the appropriate datapoints
            selection = complete_benchmark_data["Paths"]<=benchmark_datapoint["Paths"]
            temp_model = generate_latency_model(complete_benchmark_data[selection]["Paths"],complete_benchmark_data[selection]["Latency"])
            temp_model_latency = temp_model(fixed_runtime_paths)
            
            #calculating the errors
            error = fixed_runtime_latency - temp_model_latency + 0.0001
            task_relative_error[t,i] = numpy.copy(numpy.abs(error)/fixed_runtime_latency)
            task_absolute_error[t,i] = numpy.copy(error)
            
    #Taking the mean of all tasks for a particular platform
    benchmark_runtime_paths_ratio[p] = numpy.mean(task_benchmark_runtime_paths_ratio,axis=0)
    benchmark_runtime_latency_ratio[p] = numpy.mean(task_benchmark_runtime_latency_ratio,axis=0)
    relative_latency_error[p] = scipy.stats.gmean(task_relative_error,axis=0)
    absolute_latency_error[p] = numpy.mean(task_absolute_error,axis=0)
```

Calculating the relative and absolute errors for the latency model for a fixed benchmarking time.

```python
platforms_interest = [(platform,platform_type) for platform,platform_type in platforms[:]] #if ("phi" not in platform and "opencl_gpu" not in platform_type)
tasks_interest = option_tasks[:]

min_benchmarking_points = 3
datapoints = 5

benchmark_runtime_paths_ratio = numpy.empty((len(platforms_interest),datapoints))
benchmark_runtime_latency_ratio = numpy.empty_like(benchmark_runtime_paths_ratio)
relative_latency_error = numpy.empty_like(benchmark_runtime_paths_ratio)
absolute_latency_error = numpy.empty_like(benchmark_runtime_paths_ratio)

for p,(platform,platform_type) in enumerate(platforms_interest):
    task_benchmark_runtime_paths_ratio = numpy.empty((len(tasks_interest),datapoints))
    task_benchmark_runtime_latency_ratio = numpy.empty_like(task_benchmark_runtime_paths_ratio)
    task_relative_error = numpy.empty_like(task_benchmark_runtime_paths_ratio)
    task_absolute_error = numpy.empty_like(task_benchmark_runtime_paths_ratio)
    
    benchmarking_time = 600.0/128
    
    if(platform=="pi"): min_benchmarking_points = 2
    
    
    for t in range(len(tasks_interest)):
        task = str(t)
        platform_extra_time = 0.0
        
        #selecting benchmark data and generating latency model from it
        benchmarking_data = data_select(data,platform,platform_type,task,benchmarking_time)
        
        #routine for increasing benchmarking time if necessary
        if(len(benchmarking_data)<min_benchmarking_points):
            temp_benchmarking_time = benchmarking_time
            while(len(benchmarking_data)<min_benchmarking_points):
                temp_benchmarking_time += 1.0 #extra second
                platform_extra_time += 1.0
                benchmarking_data = data_select(data,platform,platform_type,task,temp_benchmarking_time)
                
            #print("%d task on %s(%s) required %f"%(t,platform,platform_type,temp_benchmarking_time))
            #raise Exception("Insufficient Benchmarking data for %d task on %s(%s)"%(t,platform,platform_type))
        
        latency_model = generate_latency_model(benchmarking_data["Paths"],benchmarking_data["Latency"])
        
        fixed_benchmark_paths = benchmarking_data["Paths"][-1]
        fixed_benchmark_latency = benchmarking_data["Latency"][-1]
        
        #selecting verification data
        verification_data = data_select(data,platform,platform_type,task)
        verification_selection = verification_data["Paths"]>=fixed_benchmark_paths
        verification_data = verification_data[verification_selection][:datapoints]
        if(len(verification_data)<datapoints): raise Exception("Insufficient verification data for task %d on %s(%s)"%(t,platform,platform_type))
        
        verification_paths = verification_data["Paths"]
        task_benchmark_runtime_paths_ratio[t] = verification_paths/fixed_benchmark_paths
        verification_latencies = verification_data["Latency"]
        task_benchmark_runtime_latency_ratio[t] = verification_paths/fixed_benchmark_latency
        
        #generating latency predicitions
        latency_predictions = numpy.array(map(latency_model,verification_paths))
        
        #calculating task error
        task_absolute_error[t] = verification_latencies - latency_predictions
        task_relative_error[t] = numpy.abs(task_absolute_error[t])/verification_latencies + 0.0001
    
        if(platform_extra_time): print("%s(%s) required %fS extra for task %s"%(platform,platform_type,platform_extra_time,task))
    
    #finding platform means
    benchmark_runtime_paths_ratio[p] = numpy.mean(task_benchmark_runtime_paths_ratio,axis=0)
    benchmark_runtime_latency_ratio[p] = numpy.mean(task_benchmark_runtime_latency_ratio,axis=0)
    relative_latency_error[p] = scipy.stats.gmean(task_relative_error,axis=0)
    absolute_latency_error[p] = numpy.mean(task_absolute_error,axis=0)
```

##### Accuracy Model

```python
def sum_95CIs(paths,accuracies):
    variance = numpy.sum((paths**0.5*accuracies/1.96)**2)
    confidence = 1.96*(variance/numpy.sum(paths))**0.5
    
    return confidence
```

Calculating the relative and absolute accuracy models. The error is the actual 95% confidence interval measured versus what is the 95% confidence interval predicted by the model for that number of simulations. An error > 0 indicates that the model has *overestimated* the number of simulations required to achieve that level of accuracy, while an error < 0 indicates that the model has *underestimated* the simulations required.

The relative error performance will be limited by any inherent bias in the model.

```python
platforms_interest = [(platform,platform_type) for platform,platform_type in platforms[:]] #if ("phi" not in platform and "opencl_gpu" not in platform_type)
tasks_interest = option_tasks[:]
datapoints = 4

benchmark_runtime_paths_ratio = numpy.empty((len(tasks_interest),datapoints))
benchmark_runtime_latency_ratio = numpy.empty_like(benchmark_runtime_paths_ratio)
relative_accuracy_error = numpy.empty_like(benchmark_runtime_paths_ratio)
absolute_accuracy_error = numpy.empty_like(benchmark_runtime_paths_ratio)
offsets = []
for t in range(len(tasks_interest)):
    task = str(t)
    
    platform_benchmark_runtime_paths_ratio = numpy.empty((len(platforms_interest),datapoints))
    platform_benchmark_runtime_latency_ratio = numpy.empty_like(platform_benchmark_runtime_paths_ratio)
    platform_relative_error = numpy.empty_like(platform_benchmark_runtime_paths_ratio)
    platform_absolute_error = numpy.empty_like(platform_benchmark_runtime_paths_ratio)
    for p,(platform,platform_type) in enumerate(platforms_interest):
        task_data = data_select(data,platform,platform_type,task)[:datapoints+1]
        if(numpy.any(task_data["Accuracy"]==0)): temp_accuracy_model = lambda x: numpy.inf
        else: temp_accuracy_model = generate_accuracy_model(task_data["Paths"],task_data["Accuracy"])
            
        #Making sure the benchmark data doesn't contain any zeros or the model produces finite coefficients
        offset = 0
        if(not numpy.isfinite(temp_accuracy_model(1))):
            offset = 1
            while(not numpy.isfinite(temp_accuracy_model(1))):
                task_data = data_select(data,platform,platform_type,task)[offset:offset+datapoints+1]
                if(numpy.any(task_data["Accuracy"]==0)):temp_accuracy_model = lambda x: numpy.inf
                else: temp_accuracy_model = generate_accuracy_model(task_data["Paths"],task_data["Accuracy"])
                offset += 1
            
        offsets += [offset]
        if(len(task_data)<datapoints): raise Exception("Insufficient datapoints for task %d on %s(%s)"%(t,platform,platform_type))
        runtime_paths = task_data["Paths"][-1]
        runtime_latency = task_data["Latency"][-1]
        runtime_accuracy = task_data["Accuracy"][-1]
        
        #increasing the amount of benchmarking datapoints - starting off with at least 2
        for i,datapoint in enumerate(task_data[1:]):
            selection = task_data["Paths"]<=datapoint["Paths"]
            temp_data = task_data[selection]
            
            platform_benchmark_runtime_paths_ratio[p,i] = datapoint["Paths"]/runtime_paths
            platform_benchmark_runtime_latency_ratio[p,i] = datapoint["Latency"]/runtime_latency
            
            temp_accuracy_model = generate_accuracy_model(temp_data["Paths"],temp_data["Accuracy"])
            
            platform_absolute_error[p,i] = runtime_accuracy - temp_accuracy_model(runtime_paths)
            platform_relative_error[p,i] = platform_absolute_error[p,i]/runtime_accuracy
            
    benchmark_runtime_paths_ratio[t] = numpy.mean(platform_benchmark_runtime_paths_ratio,axis=0)
    benchmark_runtime_latency_ratio[t] = numpy.mean(platform_benchmark_runtime_latency_ratio,axis=0)
    relative_accuracy_error[t] = numpy.mean(numpy.abs(platform_relative_error),axis=0)
    absolute_accuracy_error[t] = numpy.mean(platform_absolute_error,axis=0)
    
sanity_benchmark_paths_ratio = numpy.copy(benchmark_runtime_paths_ratio)
sanity_relative_accuracy_error = numpy.copy(relative_accuracy_error)
```
