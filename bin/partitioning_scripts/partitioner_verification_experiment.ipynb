# Partitioner Data Generation

Aim: 
* Use the predictive models to produce latency projections for a reasonable accuracy region (\$10,\$1,\$0.1,\$0.01)
* Use these latency projections to generate three partitions:
 * Simple inverse arithmetic (proportional)
 * Machine learning-based (simulated annealling with Nelder-Mead polishing)
 * SCIP-based
* Run these partitions on the set of platforms

## Setup

Importing necessary libraries

## Method

Model Verification Procedure:
1. Create $$Path(Accuracy)$$ and $$Latency(Accuracy)$$ models for each task upon each platform
2. Generate Task Partitions for the accuracy range of interest for all tasks using all of the platforms

### Model Functions

#### Paths(Accuracy) Model

$$ A = \frac{\alpha}{\sqrt{N}} $$
$$ \therefore N = (\frac{\alpha}{A})^2 $$

```python
def paths_accuracy(A,accuracy_coefficient):
    return (accuracy_coefficient/A)**2
```

#### Latency(Paths) Model

$L(N) = \beta_{prop}N + \beta_{setup}$

```python
def latency_paths(N,latency_coefficient,latency_setup):
    return latency_coefficient*N + latency_setup*(N>0)
```

#### Latency(Accuracy) Model

$$L(A) = \beta_{prop}(\frac{\alpha}{A})^2 + \beta_{setup}$$

```python
def latency_accuracy(A,accuracy_coefficient,latency_coefficient,latency_setup):
    return latency_coefficient*paths_accuracy(A,accuracy_coefficient) + latency_setup
```

#### Latency(Accuracy) Coefficient Model

$$ L_{coeff}(A) = \beta_{prop}(\frac{\alpha}{A})^2 $$

```python
def latency_accuracy_prop(A,accuracy_coefficient,latency_coefficient):
    return latency_coefficient*paths_accuracy(A,accuracy_coefficient)
```

### Cost Models

```python
def select_coefficients(data,platform,platform_type,task):
    selection = numpy.logical_and(data["Platform"]==platform,data["Type"]==platform_type)
    selection = numpy.logical_and(selection[:],data["Task"]==task)
    accuracy_coefficient,latency_coefficient,latency_setup = data[selection]["Accuracy Coefficient"],data[selection]["Latency Coefficient"],data[selection]["Latency Setup"]
    
    return accuracy_coefficient,latency_coefficient,latency_setup
```

```python
cost_matrices = []
for accuracy_target in accuracy_range[:]:
    pp_latencies = []
    setup_latencies = []
    for platform,platform_type in platforms:
        pp_latencies += [[]]
        setup_latencies += [[]]
        for t in range(len(tasks)):
            accuracy_coefficient,latency_coefficient,latency_setup = select_coefficients(data,platform,platform_type,t)
            pp_latencies[-1] += [latency_accuracy_prop(accuracy_target,accuracy_coefficient,latency_coefficient)]
            setup_latencies[-1] += [latency_setup]
            
    pp_latencies = numpy.copy(numpy.array(pp_latencies).reshape(len(platforms),len(tasks))) 
    setup_latencies = numpy.copy(numpy.array(setup_latencies).reshape(len(platforms),len(tasks))) 
    
    cost_matrices += [(setup_latencies,pp_latencies)]
```

Function for calculating the value of an allocation, given its cost matrices:

```python
def allocation_cost(allocation,pp_costs,setup_costs,number_platforms=None,number_tasks=None):
    #print(allocation)
    #print(pp_costs)
    #print(setup_costs)
    #print("PP Cost: ",numpy.any(numpy.logical_not(numpy.isfinite(pp_costs)))))
    #print("Setup Cost: ",numpy.any(numpy.logical_not(numpy.isfinite(setup_costs))))
    #print("Allocation Cost: ",numpy.any(numpy.logical_not(numpy.isfinite(allocation))))
    
    if(number_platforms and number_tasks):
        allocation = allocation.reshape(number_platforms,number_tasks)
        pp_costs = pp_costs.reshape(number_platforms,number_tasks)
        setup_costs = setup_costs.reshape(number_platforms,number_tasks)
    
    selection = numpy.isfinite(pp_costs)
    selection &= numpy.isfinite(setup_costs)
    selection = numpy.logical_not(selection)
    
    if(numpy.any(allocation[selection]>0.0)):
        invalid_values = []
        for p in range(pp_costs.shape[0]):
            for t in range(pp_costs.shape[1]):
                if (not numpy.isfinite(selection[p,t]) and allocation[p,t]>0.0): invalid_values += [(p,t)]
                    
        raise ValueError("Allocating to invalid value(s): %s"%str(invalid_values))
    
    platform_task_costs = numpy.zeros(pp_costs.shape)#pp_costs*allocation+setup_costs*(allocation>0)
    
    #for p in range(pp_costs.shape[0]):
        #for t in range(pp_costs.shape[1]):
            #if(allocation[p,t]):
                #platform_task_costs[p,t] = pp_costs[p,t]*allocation[p,t] + setup_costs[p,t]
                
    platform_task_costs = pp_costs*allocation + setup_costs*(allocation>0)
            
    """
    platform_task_costs = platform_allocation_totals
    selection = numpy.isfinite(platform_allocation_totals)
    
    platform_task_costs = numpy.zeros(pp_costs.shape)
    for p in range(setup_costs.shape[0]):
        for t in range(setup_costs.shape[1]):
            if(selection[p,t]): platform_task_costs[p,t] = platform_allocation_totals[p,t]
    """
    
    cost = numpy.max(numpy.sum(platform_task_costs,axis=1)) #find the longest running platform

    return cost
```

```python
def platform_execute(platform,option_tasks,platform_paths,accuracy_targets,queue,output_queue,error_queue,expected_times,):
    import subprocess32 as subprocess
    import time
    for i,o in enumerate(option_tasks):
        results = []
        
        for target_accuracy,paths in zip(accuracy_targets,platform_paths[:,i]):
            retries = 10
            
            accuracy_index = accuracy_targets.index(target_accuracy)
            timeout = max(expected_times[accuracy_index,i]*2,100) #give it 100% leeway or 100 seconds
            
            while(retries):
                if(paths):
                    if not(isinstance(platform,MulticoreCPU.MulticoreCPU)): #rounding up the number of paths
                        paths = int(numpy.ceil(1.0*paths/20)*20)
                    #else:
                        #paths = int(numpy.ceil(1.0*paths/10)*10)
                    
                    if(isinstance(platform,MulticoreCPU.MulticoreCPU)): 
                        mc_solver = MulticoreCPU_MonteCarlo.MulticoreCPU_MonteCarlo([o],paths,platform)
                    elif(isinstance(platform,OpenCLGPU.OpenCLGPU) and platform.device_type==8): 
                        mc_solver = OpenCLGPU_MonteCarlo.OpenCLGPU_MonteCarlo([o],paths,platform,kernel_path_max=1)
                    elif(isinstance(platform,OpenCLGPU.OpenCLGPU)): 
                        mc_solver = OpenCLGPU_MonteCarlo.OpenCLGPU_MonteCarlo([o],paths,platform)
                    elif(isinstance(platform,OpenCLAlteraFPGA.OpenCLAlteraFPGA)): 
                        mc_solver = OpenCLAlteraFPGA_MonteCarlo.OpenCLAlteraFPGA_MonteCarlo([o],paths,platform)
                    elif(isinstance(platform,MaxelerFPGA.MaxelerFPGA)):
                        if("heston" in o.underlying[0].name): mc_solver = MaxelerFPGA_MonteCarlo.MaxelerFPGA_MonteCarlo([o],1000,platform,instances=8)
                        elif("black_scholes" in o.underlying[0].name): mc_solver = MaxelerFPGA_MonteCarlo.MaxelerFPGA_MonteCarlo([o],1000,platform,instances=16)
                    
                    if(isinstance(platform,MaxelerFPGA.MaxelerFPGA)):
                        mc_solver.dummy_run()
                        mc_solver.execute()
                        
                        if("heston" in o.underlying[0].name): mc_solver = MaxelerFPGA_MonteCarlo.MaxelerFPGA_MonteCarlo([o],paths,platform,instances=8)
                        elif("black_scholes" in o.underlying[0].name): mc_solver = MaxelerFPGA_MonteCarlo.MaxelerFPGA_MonteCarlo([o],paths,platform,instances=16)
                    
              
                    try:
                        if(platform.ssh_alias!="phi" and platform.ssh_alias!=""): timeout = None
                        result = mc_solver.execute(timeout=timeout)
                        #print(expected_times[accuracy_index,i],timeout)
                        #print(i,target_accuracy,result)
                        if(numpy.isnan(float(result[1])) or float(result[1])==0): raise ValueError()
                            
                        retries = 0
                        results.append((paths,result[-1]/10e6,target_accuracy,float(result[1])))
                        output_queue.put_nowait((platform.ssh_alias,platform.name,(paths,result[-1]/10e6,target_accuracy,float(result[1]))))
                    
                    #print((platform.ssh_alias,platform.name,(paths,result[-1]/10e6,float(result[1]))))
            
                    except subprocess.CalledProcessError as e:
                        output_queue.put_nowait("Command Error!\nCommand:" + str(e.cmd) + "\nReturn: " + str(e.output) + "\nReturncode: " + str(e.returncode))
                        retries -= 1
                        
                        if not(isinstance(platform,MulticoreCPU.MulticoreCPU)): #increasing the number of paths
                            paths += 20
                        
                        """
                        else:
                            paths += 10
                        """
                        
                        if not(retries): error_queue.put_nowait((platform.ssh_alias,platform.name,paths,i,target_accuracy))
                    
                    except subprocess.TimeoutExpired as e:
                        #print("timeout on ",i,target_accuracy)
                        output_queue.put_nowait("Timeout!\nCommand:" + str(e.cmd) + "\nReturn: " + str(e.output) + "\nTimeout (expected time): " + str(e.timeout) + "("+str(expected_times[accuracy_index,i])+")")
                        timeout += max(1,timeout*0.5) #give a second or 50% more, whichever is greater
                        retries -= 1
                        time.sleep(5) #seems to need a cool down period
                        
                        if not(retries): error_queue.put_nowait((platform.ssh_alias,platform.name,paths,i,target_accuracy))
                    
                    except ValueError:
                        output_queue.put_nowait("Value Error!\nPlatform:" + platform.ssh_alias + "\nPlatform Type:" + platform.name + "\nTask: " + str(i) + "\nValue: " + result[1])
                        retries -= 1
                        
                        #if not(isinstance(platform,MulticoreCPU.MulticoreCPU)): #increasing the number of paths
                        if(timeout): timeout = timeout*(paths+20.0)/paths
                        paths += 20
                        
                        if not(retries):
                            results.append((paths,result[-1]/10e6,target_accuracy,0.0))
                            error_queue.put_nowait((platform.ssh_alias,platform.name,paths,i,target_accuracy))
                else:
                    retries = 0
                
        if(results):
            #print(str(i) + " on " + platform.ssh_alias + "("+platform.name+")" + ": " + str(results) + "\n")
            output_queue.put_nowait(str(i) + " on " + platform.ssh_alias + "("+platform.name+")" + ": " + str(results) + "\n")
            queue.put_nowait((platform.ssh_alias,platform.name,i,results))
            
        
```

```python
import multiprocessing

queue = multiprocessing.Queue(len(platforms)*len(option_tasks)*2)
output_queue = multiprocessing.Queue()
error_queue = multiprocessing.Queue(len(platforms)*len(option_tasks))
```

```python
def run_manifest(path_manifest,platforms,execution_platforms,option_tasks,accuracy_range,queue,output_queue,error_queue,manifest_times,parallel=True):
    results = []
    if(parallel):
        try:
            pool = [multiprocessing.Process(target=platform_execute,args=(p,option_tasks,path_manifest[platforms.index((p.ssh_alias,p.name))],accuracy_range,queue,output_queue,error_queue,manifest_times[platforms.index((p.ssh_alias,p.name))],)) for j,p in enumerate(execution_platforms)]
    
            for p in pool: p.start()
        
            while(True):
                while not(output_queue.empty()): print(output_queue.get())
                while not(queue.empty()): results += [queue.get()]
        
                flag = True
                for p in pool: 
                    if(p.is_alive() and flag): flag = False
                if(flag): break
            
                time.sleep(1)
            
            for p in pool: 
                p.join()
                del p
        except KeyboardInterrupt:
            for p in pool: 
                p.terminate()
                del p
    
    else: 
        for j,p in enumerate(execution_platforms): platform_execute(p,option_tasks,path_manifest[platforms.index((p.ssh_alias,p.name))],accuracy_range,queue,output_queue,error_queue,manifest_times[platforms.index((p.ssh_alias,p.name))],)
            
    return results
        
#print(scip_manifest[j].shape)
#print(accuracy_range)
        
#for p in execution_platforms: platform_execute(p,option_tasks,scip_manifest[j],accuracy_range,queue,output_queue,)
```

```python
#verification_results[accuracy_range][task] = [[paths][latency][accuracy]]
def process_results(manifest_result_set,accuracy_range,tasks,platforms):
    verification_accuracy_results = numpy.zeros((len(accuracy_range),len(platforms),len(tasks))) #[[[] for t in tasks] for target_accuracy in accuracy_range]
    verification_path_results = numpy.zeros(verification_accuracy_results.shape)
    verification_latency_results = numpy.zeros(verification_accuracy_results.shape) #[[[] for p in platforms] for target_accuracy in accuracy_range]
    
    for result in manifest_result_set:
        platform,platform_type,task,results = result
        
        platform_index = platforms.index((platform,platform_type))
        for r in results:
            paths,latency,target_accuracy,accuracy = r
            
            accuracy_index = numpy.argmin(numpy.abs(accuracy_range-target_accuracy))
            verification_accuracy_results[accuracy_index][platform_index][task] = accuracy
            verification_path_results[accuracy_index][platform_index][task] = paths
            
            verification_latency_results[accuracy_index][platform_index][task] = latency
            
            
    return verification_accuracy_results,verification_path_results,verification_latency_results
```

$$ CI_{95\%} = 1.96\frac{\sigma}{\sqrt{N}} $$
To sum the 95% CIs, the 95% CIs have to be converted to variances and the mean variance found. The resulting variance must then be converted back to a 95% CI value, i.e. :
$$ CI_{95\%,a} + CI_{95\%,b} = 1.96\frac{\sqrt{\overline{(\sqrt{N_a}\frac{\sigma_a}{1.96})^2,(\sqrt{N_b}\frac{\sigma_b}{1.96})^2}}}{\sum{N_a,N_b}} $$

```python
def generate_verification_data(verification_results,accuracy_range,tasks,platforms):
    accuracy_results,path_results,latency_results = verification_results
    
    verification_accuracies = numpy.zeros((len(accuracy_range),len(tasks)))
    #verification_paths = numpy.zeros(verification_accuracies.shape)
    verification_latencies = numpy.zeros((len(accuracy_range),len(platforms)))
    
    missing_tasks = []
    for i,accuracy_set in enumerate(accuracy_range):
        #Find the Accuracy for each task
        missing_tasks += [[]]
        for j in range(len(tasks)):
            combined_accuracy = 1.96*(numpy.sum((accuracy_results[i,:,j]*path_results[i,:,j]**0.5/1.96)**2)/numpy.sum(accuracy_results[i,:,j]>0)/numpy.sum(path_results[i,:,j]))**0.5
            if (numpy.isfinite(combined_accuracy)): verification_accuracies[i,j] = combined_accuracy
            else:
                missing_tasks[-1] += [j]
        
        #print(verification_accuracies[i,:])
        
        #Find the latency for each platform, minus the latencies from the missing tasks
        for j in range(len(platforms)):
            verification_latencies[i,j] = numpy.sum(latency_results[i,j])
            for mt in missing_tasks[-1]: verification_latencies[i,j] -= latency_results[i,j,mt]
            
    return verification_latencies,verification_accuracies,missing_tasks


            
            
```

```python
def calculate_verification_stats(verification_latencies,verification_accuracies):
    latency_results = []
    accuracy_results = []
    for vl,va in zip(verification_latencies,verification_accuracies):
        selection = numpy.logical_not(va==0) #numpy.ones(va.shape).astype(numpy.bool) 
        
        accuracy_mean = scipy.stats.gmean(va[selection])
        accuracy_std = numpy.std(va[selection])
        accuracy_max = numpy.max(va[selection])
        accuracy_min = numpy.min(va[selection])
        accuracy_results += [(accuracy_min,accuracy_mean,accuracy_std,accuracy_max)]
        
        selection =  numpy.logical_not(vl==0) #numpy.ones(vl.shape).astype(numpy.bool)
        latency_mean = scipy.stats.gmean(vl[selection])
        latency_std = numpy.std(vl[selection])
        latency_max = numpy.max(vl[selection])
        latency_min = numpy.min(vl[selection])
        latency_results += [(latency_min,latency_mean,latency_std,latency_max)]
        
    return numpy.array(latency_results),numpy.array(accuracy_results)
        
    
```
