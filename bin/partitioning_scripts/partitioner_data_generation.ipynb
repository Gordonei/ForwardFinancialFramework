# Partitioner Data Generation

Aim: 
* Use the predictive models to produce latency projections for a reasonable accuracy region (\$1,\$0.1,\$0.01,\$0.001,\$0.0001)
* Use these latency projections to generate three partitions:
 * Simple inverse arithmetic
 * Machine learning-based
 * SCIP-based
* Run these partitions on the set of platforms

## Setup

Importing necessary libraries

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy
import scipy
import scipy.optimize
import csv,time,copy
```

## Method

Model Verification Procedure:
1. Create $$Path(Accuracy)$$ and $$Latency(Accuracy)$$ models for each task upon each platform
2. Generate Task Partitions for the accuracy range of interest for all tasks using all of the platforms

## Generating Models

### Reading in Model Coefficients

Reading model coefficient data

```python
with open('model_coefficients_complete2.csv','r') as benchmark_file:
    csvreader = csv.reader(benchmark_file,delimiter=',')
    
    headings = csvreader.next()
    heading_tuples = [(h,"a30") for h in headings if h in ["Platform","Type"]]
    heading_tuples += [(h,"f8") for h in headings if h not in ["Platform","Type"]]
    
    data = []
    for row in csvreader: data.append(tuple(row))
            
    data = numpy.array(data,dtype=heading_tuples)
```

Generating the list of platforms

```python
platforms_set = set(data["Platform"])
platform_types_set = set(data["Type"])

platforms = []
for platform in platforms_set:
    for platform_type in platform_types_set:
        selection = numpy.logical_and(data["Platform"]==platform,data["Type"]==platform_type)
        if(numpy.any(selection)): platforms.append((platform,platform_type))
            
platforms = [(platform,platform_type) for platform,platform_type in platforms] #if not(platform=="phi" and platform_type=="opencl_gpu")

for platform in platforms: print(platform)
```

list of task identifiers:

```python
tasks = sorted(list(set(data["Task"])))
```

### Model Functions

#### Paths(Accuracy) Model

```python
def paths_accuracy(A,accuracy_coefficient):
    return (accuracy_coefficient/A)**2
```

#### Latency(Accuracy) Model

```python
def latency_accuracy(accuracy_coefficient,latency_coefficient,latency_setup):
    return lambda A,latency_coefficient=latency_coefficient,accuracy_coefficient=accuracy_coefficient: latency_coefficient*paths_accuracy(A,accuracy_coefficient) + latency_setup
```

#### Latency(Accuracy) Coefficient Model

```python
def latency_accuracy_coefficient(accuracy_coefficient,latency_coefficient):
    return lambda A,latency_coefficient=latency_coefficient,accuracy_coefficient=accuracy_coefficient: latency_coefficient*paths_accuracy(A,accuracy_coefficient)
```

## Generating Partitions

### Setup

Accuracy range of interest, with logarithmic stepping:

```python
"""
accuracy_range = range(10,1,-1) #1
accuracy_range += list(numpy.arange(1,0.1,-0.1)) #0.1
accuracy_range += list(numpy.arange(0.1,0.01,-0.01)) #0.01
accuracy_range += list(numpy.arange(0.01,0.001,-0.001)) #0.001
accuracy_range += [0.001]
"""
accuracy_range = [10,1,0.1,0.01,0.001]

accuracy_range = numpy.array(accuracy_range)
print(accuracy_range)
```

Function for selection the coefficients for a particular platform-task combination from the coefficient data array:

```python
def select_coefficients(data,platform,platform_type,task):
    selection = numpy.logical_and(data["Platform"]==platform,data["Type"]==platform_type)
    selection = numpy.logical_and(selection[:],data["Task"]==task)
    accuracy_coefficient,latency_coefficient,latency_setup = data[selection]["Accuracy Coefficient"],data[selection]["Latency Coefficient"],data[selection]["Latency Setup"]
    
    return accuracy_coefficient,latency_coefficient,latency_setup
```

Generating the cost models that are used by all of the partitioners:

```python
cost_models = []
for platform,platform_type in platforms:
    cost_models += [[]]
    for task in tasks[:]:
        accuracy_coefficient,latency_coefficient,latency_setup = select_coefficients(data,platform,platform_type,task)
        coefficient_model = latency_accuracy_coefficient(accuracy_coefficient,latency_coefficient)
        setup_model = lambda A,latency_setup=latency_setup: latency_setup
        cost_models[-1] += [(coefficient_model,setup_model)]
```

Using the cost models to generate both the per-path latency matrices as well as the setup latency matrices for the accuracy range of interest

```python
cost_matrices = []
for accuracy_target in accuracy_range[:]:
    pp_latencies = []
    setup_latencies = []
    for platform_models in cost_models:
        pp_latencies += [[]]
        setup_latencies += [[]]
        for task_pp_model,task_setup_model in platform_models:
            pp_latencies[-1] += [task_pp_model(accuracy_target)]
            setup_latencies[-1] += [task_setup_model(accuracy_target)]
     
    pp_latencies = numpy.copy(numpy.array(pp_latencies).reshape(len(platforms),len(tasks[:])))
    setup_latencies = numpy.copy(numpy.array(setup_latencies).reshape(len(platforms),len(tasks[:])))
    
    cost_matrices += [(setup_latencies,pp_latencies)]
```

### Proportional Partitioner

In generating the proportional partition, the relative platform latencies (i.e. the sum of all of the tasks upon that platform) are used used to partition the workload. This is given by $$ \vec{A} = \frac{\frac{1}{\hat{L}_p}}{\sum\nolimits_{k \in \mu}{\frac{1}{\hat{L}_k}}}$$, where $$\vec{L}_p = \sum\nolimits_{t \in \tau}{\bf{ETC_{pp,p,t}}+\bf{ETC_{setup,p,t}}} \:\: p \in \mu$$

An important caveat is that the sum of only those tasks of which exist across all platforms are used to do this relative ranking.

From a technical point of view, the proportional partitioning function only requires the sum of the setup and pp latency matrices, but I've kept the interface consistent for the later solvers which require the separate matrices.

```python
def proportional_partition(setup_costs,pp_costs):
    #selecting the tasks that will be used to build the proportional schedule
    temp_latencies = numpy.sum(setup_costs + pp_costs,axis=0) #finding the task totals across all platforms
    selection = numpy.logical_not(numpy.isinf(temp_latencies)) #only selecting those tasks which have values across all platforms

    latencies = numpy.sum(setup_costs[:,selection] + pp_costs[:,selection],axis=1)
    allocation = numpy.linalg.norm(latencies)/latencies/numpy.sum(numpy.linalg.norm(latencies)/latencies) #numpy.sum(latencies)/latencies/numpy.sum(numpy.sum(latencies)/latencies)
    
    return allocation
```

```python
proportional_allocations = []
for accuracy_target,(setup_cost_matrix,pp_cost_matrix) in zip(accuracy_range,cost_matrices):
    proportional_allocations += [proportional_partition(setup_cost_matrix,pp_cost_matrix).reshape(len(platforms),1)]
```

### Numerical Optimiser-based Partitioner

```python
def allocation_cost(allocation,pp_costs,setup_costs):
    platforms = setup_costs.shape[0]
    tasks = setup_costs.shape[1]
    
    #platform_allocation_totals = []
    #for p in range(platforms):
    platform_allocation_totals = numpy.sum(pp_costs*allocation+setup_costs,axis=1)
    selection = numpy.logical_not(numpy.isinf(platform_allocation_totals)) #avoiding those tasks which don't have sensible coefficents
    
    #print("platform allocation totals:",platform_allocation_totals)
    cost = numpy.max(platform_allocation_totals[selection]) #find the longest running platform
    #if(numpy.isinf(cost)): cost = numpy.finfo(numpy.float64).max
    
    return cost
```

```python
class SumValueError(Exception):
    def __init__(self, value):
        self.value = value
    

class BoundValueError(Exception):
    def __init__(self, value):
        self.value = value
```

```python
def constrain_solution(x,platforms,tasks):
    x_new = numpy.copy(numpy.abs(x))

    for t in range(tasks):
        x_new[:,t] = numpy.copy(x_new[:,t])/numpy.sum(x_new[:,t])
        if(numpy.any(0.0>=numpy.sum(x_new[:,t])>=1.0)): raise BoundValueError(t)
        if(numpy.abs((numpy.sum(x_new[:,t])-1))>0.0001): raise SumValueError(t)
            
    return x_new
```

```python
def optimiser_solution(setup_costs,pp_costs,platforms,tasks,timeout=10):
    #platforms = setup_costs.shape[0]
    #tasks = setup_costs.shape[1]
    
    def objective_function(x,setup_costs,pp_costs):
        #platforms,tasks = pp_costs.shape

        if(x.size!=(len(platforms)*len(tasks))): x = x[0]
            
        x = x.reshape(len(platforms),len(tasks))
        try: x = constrain_solution(x,len(platforms),len(tasks))
        except BoundValueError as e:
            print(int(e.value))
            for index,j in enumerate(x[:,int(e.value)]):
                if(0>j>1.0): print(j)
            #sys.exit(1)
                    
        except SumError as e:
            print(int(e.value))
            print(numpy.sum(x[:,e.value]))
            #sys.exit(1)
        
        return allocation_cost(x,pp_costs,setup_costs)
    
    #Starts off using the proportional allocation as a guess
    platform_allocation = proportional_partition(setup_costs,pp_costs).reshape(len(platforms),1)
    #platform_allocation = numpy.ones(len(platforms))*1.0/len(platforms)
    #print("initial allocation total: ",numpy.sum(platform_allocation.T))
    
    initial_allocation = numpy.ones(setup_costs.shape)
    for p in range(len(platforms)):
        initial_allocation[p,:] = platform_allocation[p]
        
    #print(initial_allocation)
    #initial_allocation_value = allocation_cost(initial_allocation,pp_costs,setup_costs)
        
    sa_maxiters = 10
    nm_maxiters = 100
    sa_nm_ratio = 0.9
    
    flag = True
    duration = 0.0
    T0 = 100*len(platforms)*len(tasks)
    while(flag):
        flag = False
        
        sa_start_time = time.time()
        sa_res = scipy.optimize.minimize(objective_function,initial_allocation.reshape(len(platforms)*len(tasks),1),method='Anneal',args=(setup_costs,pp_costs,),options={"maxiter":sa_maxiters,'lower':0.0,'upper':1.0,'T0':T0}) #Simulated Annealing
        #sa_res = scipy.optimize.basinhopping(objective_function,initial_allocation.reshape(platforms*tasks,1),niter=sa_maxiters,minimizer_kwargs={"args":(setup_costs,pp_costs,)},stepsize=0.1,disp=True)
        sa_end_time = time.time()
        
        #print(sa_res)
        
        T0 = sa_res.T
        sa_duration = sa_end_time-sa_start_time
        if(sa_res.x.size != len(platforms)*len(tasks)): sa_res.x = sa_res.x[0]
        output_allocation = sa_res.x.reshape(len(platforms),len(tasks)) #constrain_solution(,platforms,tasks)

        duration += sa_duration
        initial_allocation = numpy.copy(output_allocation)
        
        if(duration<timeout*sa_nm_ratio and not sa_res.success): flag = True # or not nm_res.success
        #if(flag): print("loop repeat: sa=",sa_res.success," sa_maxiters=",sa_maxiters," sa_duration=",sa_duration," duration=",duration) #," nm=",nm_res.success," nm_maxiters=",nm_maxiters, " nm_duration=",nm_duration
        #else: print("loop exit: sa=",sa_res.success," sa_maxiters=",sa_maxiters," sa_duration=",sa_duration,"sa_result=",sa_res.fun," duration=",duration) #," nm=",nm_res.success," nm_maxiters=",nm_maxiters, " nm_duration=",nm_duration
        
    #print(output_allocation)
    initial_allocation = constrain_solution(numpy.copy(output_allocation),len(platforms),len(tasks))
    #print(initial_allocation)
    
    #"Polishing" using the Simplex algorithm
    flag = True
    while(flag):
        flag = False
        nm_start_time = time.time()
        nm_res = scipy.optimize.minimize(objective_function,initial_allocation.reshape(len(platforms)*len(tasks),1),method='Nelder-Mead',args=(setup_costs,pp_costs,),options={"maxiter":nm_maxiters}) 
        nm_end_time = time.time()
        
        #print(nm_res)
        
        nm_duration = nm_end_time-nm_start_time
    
        duration += nm_duration
        initial_allocation = numpy.copy(nm_res.x.reshape(len(platforms),len(tasks)))
        
        if(duration<timeout and not nm_res.success): flag = True
            
    output_allocation = constrain_solution(initial_allocation.reshape(len(platforms),len(tasks)),len(platforms),len(tasks))
    output_allocation_value = allocation_cost(output_allocation,pp_costs,setup_costs)
    
    print("Time: ",duration)
    
    temp_proportional_allocation = proportional_partition(setup_costs,pp_costs).reshape(len(platforms),1)
    
    proportional_allocation = numpy.ones(setup_costs.shape)
    for p in range(len(platforms)): proportional_allocation[p,:] = temp_proportional_allocation[p]
        
    proportional_allocation_value = allocation_cost(proportional_allocation,pp_costs,setup_costs)
    
    print("Projected Improvement",proportional_allocation_value/output_allocation_value)
    
    return output_allocation
```

```python
numerical_optimiser_allocations = []
for accuracy_target,(setup_cost_matrix,pp_cost_matrix) in zip(accuracy_range[:],cost_matrices[:]):
    numerical_optimiser_allocations += [optimiser_solution(setup_cost_matrix,pp_cost_matrix,platforms,tasks,300)]
    #print(numerical_optimiser_allocations[-1])
    print(numerical_optimiser_allocations[-1].shape)
```

### SCIP-based Partitioner

```python
import subprocess32 as subprocess
import os

def scip_partitioner(setup_costs,pp_costs,platforms,tasks,timeout=10,acc_id=None):
    #platforms = setup_costs.shape[0]
    #tasks = setup_costs.shape[1]
    
    if(acc_id==None): acc_id = time.time()
    filename_suffix = "%d_%d_%d_%f"%(len(platforms),len(tasks),timeout,acc_id)
    filename_suffix = filename_suffix.replace('.','-')
    
    platform_filename = "platforms_%s.txt"%filename_suffix
    task_filename = "tasks_%s.txt"%filename_suffix
    setup_filename = "latency_setup_%s.txt"%filename_suffix
    pp_filename = "latency_pp_%s.txt"%filename_suffix
    settings_filename = "settings_%s.set"%filename_suffix
    zimpl_output_string = "platform_schedule_%s"%filename_suffix
    
    #Writing Parameters to datafile for SCIP
    platform_file = open(platform_filename,"w")
    for tp in range(len(platforms)): platform_file.write("%s\n"%tp)
    platform_file.close()

    task_file = open(task_filename,"w")
    for tt in range(len(tasks)): task_file.write("%s\n"%tt)
    task_file.close()

    
    latencysetup_file = open(setup_filename,"w")
    latencypp_file = open(pp_filename,"w")
    
    selection = numpy.logical_not(numpy.isinf(pp_costs))
    platform_max = numpy.sum(setup_costs[selection]+pp_costs[selection])
    print("setup max:",platform_max)
    for tp in range(len(platforms)):
        for tt in range(len(tasks)):
            if (numpy.isinf(setup_costs[tp,tt]) or numpy.isinf(pp_costs[tp,tt]) or setup_costs[tp,tt]<0 or pp_costs[tp,tt]<0):
                latencysetup_file.write("%s %s %f\n"%(tp,tt,0.0)) #Setting any tasks with infinite coefficients to take longer than all of the setup times added together
                latencypp_file.write("%s %s %f\n"%(tp,tt,platform_max))
            else:
                latencysetup_file.write("%s %s %s\n"%(tp,tt,setup_costs[tp,tt])) #Just need the latency setup time which is constant
                latencypp_file.write("%s %s %s\n"%(tp,tt,pp_costs[tp,tt]))

    latencypp_file.close()
    latencysetup_file.close()

    #Creating LP description problem using ZIMPL
    zimpl_output = subprocess.check_output(["zimpl","-o",zimpl_output_string,"-D","PLATFORM_FILE=%s"%platform_filename,"-D","TASK_FILE=%s"%task_filename,"-D","LATENCY_PP_FILE=%s"%pp_filename,"-D","LATENCY_SETUP_FILE=%s"%setup_filename,"platform_schedule.zpl"])

    #Setting paramters for SCIP
    settings_file = open(settings_filename,"w")
    settings_file.write("limits/time = %f\n"%timeout)
    settings_file.close()

    #Running SCIP
    start_time = time.time()
    scip_output = subprocess.check_output(["scip","-f","%s.lp"%zimpl_output_string,"-s","%s"%settings_filename])
    end_time = time.time()

    #Parsing SCIP results
    gap_found = False
    value_found = False
    #print(scip_output)
    
    allocation = numpy.zeros((len(platforms),len(tasks)))
    #print scip_output
    for so in scip_output.split("\n"):
        if("Gap" in so and not gap_found):
            gap = so.split(":")[1].strip(' ').strip("%")
            if("infinite" in gap.lower()): gap = numpy.inf
            else: gap = float(gap)
            gap_found = True
        elif("task_max " in so and not value_found):
            value = so.split(":")[0].split('(')[0].split(' ')[-2]
            value_found = True
        elif("A$" in so and "task_max" not in so):
            temp_list = so.split(" ")
            try: a_value = float(temp_list[-2])
            except ValueError: 
                print("Error caused by:",temp_list[-2])
                a_value = 0.0
            dummy,platform,task = temp_list[0].split("$")
            #except: print("Error caused by: ",temp_list[0])
            allocation[int(platform),int(task)] = a_value
            
    print("gap:",gap)
    
    temp_proportional_allocation = proportional_partition(setup_costs,pp_costs).reshape(len(platforms),1)
    
    proportional_allocation = numpy.ones(setup_costs.shape)
    for p in range(len(platforms)): proportional_allocation[p,:] = temp_proportional_allocation[p]
        
    proportional_allocation_value = allocation_cost(proportional_allocation,pp_costs,setup_costs)
    
    print("Projected Improvement",proportional_allocation_value/float(value))
    
    os.remove(platform_filename)
    os.remove(task_filename)
    os.remove(setup_filename)
    os.remove(pp_filename)
    os.remove("%s.lp"%zimpl_output_string)
    os.remove("%s.tbl"%zimpl_output_string)
    os.remove(settings_filename)
    
    return allocation
```
